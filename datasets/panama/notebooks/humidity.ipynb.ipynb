```python
import numpy as np
import pandas as pd 
import matplotlib.pyplot as plt
import statsmodels.api as sm
```

```python
data = pd.read_csv('Panama.csv')
```

```python
y = data['nat_demand']
x1 = data['QV2M_toc']
```

```python
plt.scatter(x1,y)
plt.xlabel('Date to time', fontsize = 10)
plt.ylabel('Electricity demand', fontsize = 10)
plt.grid(True)
plt.show()
```

```python
x = sm.add_constant(x1)
results = sm.OLS(y,x).fit()
results.summary()
```

```python
plt.scatter(x1,y)
yhat = -4386.7034*x1 +1263.2005
fig = plt.plot(x1,yhat,lw = 4,c = 'orange',label = 'regression line')
plt.xlabel('SAT', fontsize = 10)
plt.ylabel('GPA', fontsize = 10)
plt.show()
```
